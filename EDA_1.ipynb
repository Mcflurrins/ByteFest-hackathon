```python
# show fields with highest % of people with postgraduate degrees
df = df_combined.copy()
mask = df['Category'].astype(str).str.lower().str.strip() != 'total'  # exclude aggregate row
df.loc[mask, 'percent_postgrad'] = df.loc[mask, 'Postgraduate Degree'] / df.loc[mask, 'Total(a)'] * 100

top_postgrad = (df.loc[mask, ['Category', 'Postgraduate Degree', 'Total(a)', 'percent_postgrad']]
                  .sort_values('percent_postgrad', ascending=False))
top_postgrad['percent_postgrad'] = top_postgrad['percent_postgrad'].round(2)

# Exclude age fields (those where Category is a number or contains '-')
age_mask = ~df.loc[mask, 'Category'].str.contains(r'^\d|–', regex=True)
top_postgrad_no_age = top_postgrad[age_mask]

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(top_postgrad_no_age['Category'].head(10), top_postgrad_no_age['percent_postgrad'].head(10))
plt.xlabel('% with Postgraduate Degree')
plt.title('Top Fields by % with Postgraduate Degree (excluding age groups)')
plt.gca().invert_yaxis()
plt.show()
```
