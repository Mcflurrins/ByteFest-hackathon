```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
input_data_1 = "/content/QAW2223DC - Table 2.csv"
df = pd.read_csv(input_data_1)

# Check the actual column names and clean them
df.columns = df.columns.str.strip()

# The first column appears to be the unnamed index column with qualification info
# Let's use the first column (index 0)
first_col = df.columns[0]

# Filter rows for Level of highest non-school qualification (rows 1-6)
qual_data = df.iloc[1:7].copy()

# Clean the 'Unemployed' column - remove commas and convert to float
qual_data['Unemployed'] = qual_data['Unemployed'].astype(str).str.replace(',', '').astype(float)

# Get qualification names from the first column
qual_names = qual_data[first_col].values

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(range(len(qual_data)), qual_data['Unemployed'], color='steelblue', edgecolor='black')
plt.xlabel('Level of Highest Non-School Qualification', fontsize=12, fontweight='bold')
plt.ylabel('Unemployed (thousands)', fontsize=12, fontweight='bold')
plt.title('Unemployment by Level of Highest Non-School Qualification', fontsize=14, fontweight='bold')
plt.xticks(range(len(qual_data)), qual_names, rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Display the data
print("\nUnemployment by Qualification Level:")
for name, unemp in zip(qual_names, qual_data['Unemployed']):
    print(f"{name}: {unemp}")
```
